```python

#Projeto de Vetores no R2

import matplotlib.pyplot as plt

x0 = 0
y0 = 0
origem = [x0, y0]

x = float(input())
y = float(input())
vetor = [x, y]

ax = plt.axes()
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
ax.quiver(origem[0],origem[1],x ,y ,angles="xy", scale_units="xy",scale=1, color='r')
ax.annotate(f'v({vetor[0]})', (vetor[0], vetor[1]), fontsize=10)
plt.plot([-4, 4], [0, 0], color='black')
plt.plot([0, 0] , [-4, 4], color='black')
ax.grid(True)
plt.show()
```
