```python
# Projeto de Vetores no R3 em Python

import matplotlib.pyplot as plt
x = 50
y = 70
z = 90

vetor = [x,y, z]
origem = [10,20, 15]

fig = plt.figure()
plt.title('Vetor 3D')
ax = plt.axes(projection='3d')
ax.set_xlim([-100, 100])
ax.set_ylim([-100, 100])
ax.set_zlim([-100, 100])
ax.quiver(origem[0], origem[1], origem[2], x,y, z, color='red')
ax.quiver(origem[0], origem[1], origem[2], -15,-50, -60, color='red')
ax.quiver(origem[0], origem[1], origem[2], 80,20, -40, color='red')
plt.savefig('figura.png')
plt.show()
```
